# Tratamiento de las variables


## Variable Retrasos (DepTimeBlk)

```python
##Separamos la columna "DepTimeBlk" en dos columnas "scheduled departure", "true output"
dfDep_ArrDelay[["scheduled departure",
                "true output",
                ]]= dfDep_ArrDelay["DepTimeBlk"].str.split("-", expand=True)
dfDep_ArrDelay
```

##Variable arrivos (ArrTimeBlk)

```python
#Separamos la columna "ArrTimeBlk" en dos columnas "scheduled arrival", "true arrival"
dfDep_ArrDelay[["scheduled arrival",
                "true arrival",
                ]]= dfDep_ArrDelay["ArrTimeBlk"].str.split("-", expand=True)
dfDep_ArrDelay
```

# Hallazgos

```python
#obtenemos el retraso total en las salidas de cada vuelo restando las variables 'scheduled departure2' - 'true output2', en la columna "Retraso salidas"
df3 ['Retraso Salidas']= df3['scheduled departure2'] - df3['true output2']
df3
```

```python
#Extraemos las columnas con las variables objeto de estudio
dfDep_ArrDelay[["scheduled arrival",
                "true arrival",
                ]]= dfDep_ArrDelay["ArrTimeBlk"].str.split("-", expand=True)
dfDep_ArrDelay
```

```python
dfDep_ArrDelay.info()
```

```python
#Transformamos las varibales de object a int64
df3["scheduled arrival2"]=df3["scheduled departure"].astype("int64")
df3
```

```python
#Transformamos las varibales de object a int64
df3["true arrival2"]=df3["true arrival"].astype("int64")
df3
```

```python
#obtenemos el retraso total en las llegadas de cada vuelo restando las variables 'scheduled arrival2' - 'true arrival2', en la columna "Retraso LLegadas"
df3 ['Retraso LLegadas']= df3['scheduled arrival2'] - df3['true arrival2']
df3
```

```python
#Obtenemos entonces el retraso total sumando las columnas 'Retraso Salidas' + "Retraso LLegadas"
df3 ['Retraso Total']= df3['Retraso Salidas'] + df3["Retraso LLegadas"]
df3
```
